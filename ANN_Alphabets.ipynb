```python
# Ravi Kiran Mahoorkar
#import library
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#load dataset
df=pd.read_csv('Alphabets_data.csv')
df
```

```python
#information
df.info()
```

```python
#describe
df.describe()
```

```python
#check na values
df.isna().sum()
```

```python
#duplicates
df.duplicated().sum()
```

```python
#drop duplicates
df.drop_duplicates(inplace=True)
df = df.reset_index(drop=True)
df
```

```python
# standarize data using standard scale
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is your DataFrame and you want to standardize the numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns

# Create a StandardScaler object
scaler = StandardScaler()

# Fit and transform the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Now 'df' contains the standardized numerical data
df
```

```python
# Perform one-hot encoding on the 'letter' column
df = pd.get_dummies(df, columns=['letter'], prefix='letter', dtype=int)

# Display the updated DataFrame
df
```

```python
# split data for train and test

from sklearn.model_selection import train_test_split
# Convert the DataFrame to a NumPy array
array = df.values

# Separate features (X) and target (y)
X = array[:,0:16]
y = array[:,16:]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Now you have X_train, X_test, y_train, and y_test
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#importing model libraries
from sklearn.model_selection import GridSearchCV, KFold
from tensorflow.keras import models, layers
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.optimizers import Adam
```

# **Model**

```python
# Model Building

# Initialize the model
model = Sequential()
# Input layer and first hidden layer
model.add(Dense(128, input_dim=16, activation='relu'))
# Second hidden layer
model.add(Dense(64, activation='relu'))
# Third hidden layer
model.add(Dense(32, activation='relu'))
# Output layer
model.add(Dense(26, activation='softmax'))
# compile the keras model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
history = model.fit(X_train, y_train, epochs=100, batch_size=32,validation_data=(X_test, y_test))
# evaluate the keras model
scores = model.evaluate(X_train, y_train)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# training accuracy
print(f"{model.metrics_names[1]}: {scores[1]*100:.2f}%")
```

```python
#Testing Accuracy
test_scores = model.evaluate(X_test, y_test)
print("Test Accuracy: %.2f%%" % (test_scores[1] * 100))
```

```python
# report

from sklearn.metrics import classification_report
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)
print(classification_report(y_true_classes, y_pred_classes))
```

```python
# summarize history for accuracy

import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

## Tuning of Hyperparameters:- batch_size and epochs

```python
# create model1
def create_model1():
    model1 = Sequential()
    model1.add(Dense(128, input_dim=16,kernel_initializer='random_uniform', activation='relu'))
    # Second hidden layer
    model1.add(Dense(64,kernel_initializer='random_uniform', activation='relu'))
    # Third hidden layer
    model1.add(Dense(32,kernel_initializer='random_uniform', activation='relu'))
    # Output layer
    model1.add(Dense(26,kernel_initializer='random_uniform', activation='softmax'))

    adam=Adam(learning_rate=0.01)
    model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model1
```

```python
# Create the model
model1 = KerasClassifier(build_fn = create_model1,verbose = 0)
# Define the grid search parameters
batch_size = [20,30,40]
epochs = [50,100]
# Make a dictionary of the grid search parameters
param_grid = dict(batch_size = batch_size,epochs = epochs)
# Build and fit the GridSearchCV
grid = GridSearchCV(estimator = model1,param_grid = param_grid,cv = KFold(),verbose = 0)
grid_result = grid.fit(X_train,y_train)
```

```python
# Summarize the results
print('Best : {}, using {}'.format(grid_result.best_score_,grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
  print('{},{} with: {}'.format(mean, stdev, param))
#Best : 0.9347570896148681, using {'batch_size': 20, 'epochs': 100}
```

## Tuning of Hyperparameters:- Learning rate and Drop out rate

```python
# Defining the model2
from keras.layers import Dropout

def create_model2(learning_rate,dropout_rate):

    model2 = Sequential()
    model2.add(Dense(128, input_dim=16,kernel_initializer='normal', activation='relu'))
    model2.add(Dropout(dropout_rate))
    # Second hidden layer
    model2.add(Dense(64,kernel_initializer='normal', activation='relu'))
    model2.add(Dropout(dropout_rate))
    # Third hidden layer
    model2.add(Dense(32,kernel_initializer='normal', activation='relu'))
    model2.add(Dropout(dropout_rate))
    # Output layer
    model2.add(Dense(26,kernel_initializer='normal', activation='softmax'))

    adam = learning_rate
    model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model2
```

```python
# Create the model

model2 = KerasClassifier(build_fn = create_model2,verbose = 0,batch_size = 20,epochs = 100)

# Define the grid search parameters

learning_rate = [0.001,0.01]
dropout_rate = [0.1,0.2]

# Make a dictionary of the grid search parameters

param_grids = dict(learning_rate = learning_rate,dropout_rate = dropout_rate)

# Build and fit the GridSearchCV

grid = GridSearchCV(estimator = model2,param_grid = param_grids,cv = KFold(),verbose = 0)
grid_result = grid.fit(X_train,y_train)
```

```python
# Summarize the results
print('Best : {}, using {}'.format(grid_result.best_score_,grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
  print('{},{} with: {}'.format(mean, stdev, param))
#Best : 0.9501882910728454, using {'dropout_rate': 0.1, 'learning_rate': 0.001}
```

# Tuning of Hyperparameters:- Activation Function and Kernel Initializer

```python
# Defining the model3

def create_model3(activation_function,init):

    model3 = Sequential()
    model3.add(Dense(128, input_dim=16,kernel_initializer = init,activation = activation_function))
    model3.add(Dropout(0.1))
    # Second hidden layer
    model3.add(Dense(64,kernel_initializer = init,activation = activation_function))
    model3.add(Dropout(0.1))
    # Third hidden layer
    model3.add(Dense(32,kernel_initializer = init,activation = activation_function))
    model3.add(Dropout(0.1))
    # Output layer
    model3.add(Dense(26, activation='softmax'))

    adam = Adam(learning_rate=0.001)
    model3.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model3
```

```python
# Create the model

model3 = KerasClassifier(build_fn = create_model3,verbose = 0,batch_size = 30,epochs = 10)

# Define the grid search parameters
activation_function = ['softmax','relu','tanh']
init = ['uniform','normal']

# Make a dictionary of the grid search parameters
param_grids = dict(activation_function = activation_function,init = init)

# Build and fit the GridSearchCV

grid = GridSearchCV(estimator = model3,param_grid = param_grids,cv = KFold(),verbose = 0)
grid_result = grid.fit(X_train,y_train)
```

```python
# Summarize the results
print('Best : {}, using {}'.format(grid_result.best_score_,grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
  print('{},{} with: {}'.format(mean, stdev, param))
#Best : 0.8629570484161377, using {'activation_function': 'relu', 'init': 'normal'}
```

```python
# Defining the model_f

def create_model_f():

    model_f = Sequential()
    model_f.add(Dense(128, input_dim=16,kernel_initializer = 'normal',activation = 'relu'))
    model_f.add(Dropout(0.1))
    # Second hidden layer
    model_f.add(Dense(128,kernel_initializer = 'normal',activation = 'relu'))
    model_f.add(Dropout(0.1))
    # Third hidden layer
    model_f.add(Dense(32,kernel_initializer = 'normal',activation = 'relu'))
    model_f.add(Dropout(0.1))
    # Output layer
    model_f.add(Dense(26, activation='softmax'))

    adam = Adam(learning_rate=0.001)
    model_f.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model_f
```
